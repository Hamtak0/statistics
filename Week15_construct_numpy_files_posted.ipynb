```python
DATA_NAME = 'fruits/'
class_list = ['apple_fruits', 'banana', 'coconut_fruits']

IMG_SIZE = 224

DATA_NAME_PATH = './' + DATA_NAME
```

### Load Dataset

```python
# Images saved in folders with the following patterns
# data_path/
#     train/
#        class0/
#           classs0_0.jpg
#           ...
#           classs0_k.jpg
#       class1/
#         classs1_0.jpg
#         ...
#         classs1_k.jpg
#       ...
#       classN/
#         classsN_0.jpg
#         ...
#         classsN_k.jpg
#     validation/
#        class0/
#        ...
#        classN/
#     test/
#        class0/
#        ...
#        classN/
# npy/
#     X_train.npy
#     y_train.npy
#     X_val.npy
#     y_val.npy
#     X_test.npy
#     y_test.npy
```

```python
DATASET_PATH = DATA_NAME + 'dataset/'
IMAGE_PATH = DATASET_PATH

NUMPY_PATH = DATA_NAME_PATH + 'npy/'

print(DATASET_PATH)
```

```python
import os
import cv2
import numpy as np
```

```python
# Read images and convert to numpy array
# with size = num_Images x IMG_SIZE x IMG_SIZE x 3 (RGB)

# Train folder
data = []
labels = []

TRAIN_PATH = IMAGE_PATH + 'train/'
classes = os.listdir(TRAIN_PATH)

for label, category in enumerate(classes):
    folder_path = os.path.join(TRAIN_PATH, category)
    for img in os.listdir(folder_path):

        img_path = os.path.join(folder_path, img)
        try:
            image = cv2.imread(img_path)                    # OpenCV loads images in BGR format
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR format to RGB format that deep-learning model requires
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

            data.append(image)
            labels.append(label)
        except:
            pass

# Convert to numpy
X_train = np.array(data)
y_train = np.array(labels)

# Normalize pixels
X_train = X_train / 255.0

print("Images shape:", X_train.shape)
print("Labels shape:", y_train.shape)

```

```python
np.save(NUMPY_PATH + "X_train.npy", X_train)
np.save(NUMPY_PATH + "y_train.npy", y_train)
```

```python
# Read images and convert to numpy array
# with size = num_Images x IMG_SIZE x IMG_SIZE x 3 (RGB)

# Validation folder
data = []
labels = []

VAL_PATH = IMAGE_PATH + 'validation/'
classes = os.listdir(VAL_PATH)

for label, category in enumerate(classes):
    folder_path = os.path.join(VAL_PATH, category)
    for img in os.listdir(folder_path):

        img_path = os.path.join(folder_path, img)
        try:
            image = cv2.imread(img_path)                    # OpenCV loads images in BGR format
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR format to RGB format that deep-learning model requires
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

            data.append(image)
            labels.append(label)
        except:
            pass

# Convert to numpy
X_val = np.array(data)
y_val = np.array(labels)

# Normalize pixels
X_val = X_val / 255.0

print("Images shape:", X_val.shape)
print("Labels shape:", y_val.shape)

np.save(NUMPY_PATH + "X_val.npy", X_val)
np.save(NUMPY_PATH + "y_val.npy", y_val)
```
