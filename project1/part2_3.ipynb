```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data

```python
Dataset_C = "worldbank_2017_2021.csv"
Dataset_B = "countries_regions.xlsx"
```

```python
orig_dfC = pd.read_csv(Dataset_C)
orig_dfC.head()
```

```python
orig_dfB = pd.read_excel(Dataset_B)
orig_dfB.rename(columns={'CountryName': 'Country Name'}, inplace=True)
orig_dfB.head()
```

```python
orig_dfm = pd.merge(orig_dfC, orig_dfB, on='Country Name', how='left')
orig_dfm.head()
```

## Clean

```python
df=orig_dfm.copy()
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
null_percent = df.isnull().sum()/len(df)*100
null_percent
```

```python
# drop column before
# a) School enrollment, primary (% gross)
# b) School enrollment, secondary (% gross)
# c) School enrollment, tertiary (% gross)
# got vipe out

target_cols = [
    'School enrollment, primary (% gross)',
    'School enrollment, secondary (% gross)',
    'School enrollment, tertiary (% gross)'
]

# Generates by gemini 3
best_thresh = 0.0
step = 0.01

for t in np.arange(0.01, 1.00, step):
    thresh_count = int(t * len(df))

    safe_to_proceed = True
    for col in target_cols:
        if df[col].count() < thresh_count:
            safe_to_proceed = False
            break
    
    if safe_to_proceed:
        best_thresh = t
    else:
        break

print(f"Highest Safe Threshold Ratio: {best_thresh:.2f}")
final_thresh_count = int(best_thresh * len(df))
df_dn = df.dropna(axis=1, thresh=final_thresh_count).copy()
print(df_dn.shape)
sns.heatmap(df_dn.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()

try:
    df_dn[target_cols]
except:
    raise "Target columns got dropped"
```

```python
# Fills numerical data

skew_limit = 0.5

numeric_cols = df_dn.select_dtypes(include=['number']).columns

for col in numeric_cols:
    skewness = df_dn[col].skew()

    if abs(skewness) > skew_limit:
        fill_val = df_dn[col].median()
        df_dn[col] = df_dn[col].fillna(fill_val)
    else:
        fill_val = df_dn[col].mean()
        df_dn[col] = df_dn[col].fillna(fill_val)
```

```python
categorical_cols = df_dn.select_dtypes(include=['object']).columns

for col in categorical_cols:
    if not df_dn[col].mode().empty:
        mode_value = df_dn[col].mode()[0]
        df_dn[col] = df_dn[col].fillna(mode_value)
```

```python
total_nans = df_dn.isnull().sum().sum()

print(f"Total missing values: {total_nans}")
```

```python
sns.heatmap(df_dn.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
# save to csv

df_dn.to_csv('DatasetC&B.csv', index=False)
```
