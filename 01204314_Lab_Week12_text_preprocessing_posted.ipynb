### Tokenization

```python
# Sentence vs. Word Tokenization
from nltk.tokenize import sent_tokenize, word_tokenize 

text = 'Hello everyone. How are you? Today is a sunny day and better than yesterday.'

sentence_tokens = sent_tokenize(text)
print("Sentence Tokenization: ")
print(sentence_tokens)

word_tokens = word_tokenize(text)
print("Word Tokenization: ")
print(word_tokens)
```

```python
orig_word_tokens = word_tokens.copy()
```

### Punctuation Removal

```python
word_tokens = [token.lower() for token in word_tokens]
print(word_tokens)
```

```python
import string
cleaned_word_tokents = [token.translate(str.maketrans('', '', string.punctuation)) for token in word_tokens]
print(cleaned_word_tokents)   
```

```python
import re
# \w = word characters in a string, \s = white space
cleaned_word_tokens2 = [re.sub(r'[^\w\s]', '', token) for token in word_tokens if re.sub(r'[^\w\s]', '', token)]  
print(cleaned_word_tokens2)
```

```python
# Use RegexpTokenizer = Tokenization + Punctuation Removal
text = 'Hello everyone. How are you? Today is a sunny day and better than yesterday.'

from nltk.tokenize import RegexpTokenizer  
model = RegexpTokenizer(r'\w+')
cleaned_word_tokens3 = model.tokenize(text)
print(cleaned_word_tokens3)
```

```python
working_word_tokens = cleaned_word_tokens2
```

### Stop Word Removal

```python
from nltk.corpus import stopwords
corpus_stop_words = set(stopwords.words('english'))
print(corpus_stop_words)
```

```python
working_word_tokens = [word for word in working_word_tokens if not word in corpus_stop_words]
print("After stop-word removal:", working_word_tokens)
```

### Lemmatization

#### Part-of-speech (POS) parameters: 
#### 'n' for nouns, 'v' for verbs, 'a' for adjectives, 'r' for adverbs, 's' for satellite (auxilliary) adjectives.

```python
# For Lemmatization
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemma_working_word_tokens = [lemmatizer.lemmatize(word) for word in working_word_tokens] 
print("Original words:", working_word_tokens)
print("Lemmas:", lemma_working_word_tokens)
```

```python
# Lemmatization example 
word_list = ['change', 'changes', 'changed', 'changing', 'changable']
lemma_list = [lemmatizer.lemmatize(word) for word in word_list]
lemma_list2 = [lemmatizer.lemmatize(word,pos='v') for word in word_list]
 
print("Original words:", word_list)
print("Lemmas:", lemma_list)
print("Lemmas with POS:", lemma_list2)
```

```python
# Lemmatization example 
word_list = ['eat', 'eats', 'ate', 'eaten', 'eating', 'edible']
lemma_list = [lemmatizer.lemmatize(word) for word in word_list]
lemma_list2 = [lemmatizer.lemmatize(word,pos='v') for word in word_list]

print("Original words:", word_list)
print("Lemmas:", lemma_list)
print("Lemmas with POS:", lemma_list2)
```

```python
# Lemmatization example 
word_list = ['university', 'universe', 'universal']
lemma_list = [lemmatizer.lemmatize(word, pos='n') for word in word_list]
 
print("Original words:", word_list)
print("Lemmas:", lemma_list)
```

```python
# Lemmatization example 
word_list =  ['see', 'seen', 'saw']
lemma_list = [lemmatizer.lemmatize(word,pos='v') for word in word_list]
 
print("Original words:", word_list)
print("Lemmas:", lemma_list)
```

```python
from nltk import pos_tag
tagged_word_tokens = pos_tag(working_word_tokens)
print("Original words:", working_word_tokens)
print("POS tagged:", tagged_word_tokens)
```

```python
# To obtain POS before running lemmatization 
# When lemmatization is executed, can use POS as value of pos parameter
from nltk.corpus import wordnet

def get_wordnet_pos(token):
    tag_dict = {'N': wordnet.NOUN,
                'V': wordnet.VERB,
                'J': wordnet.ADJ,
                'R': wordnet.ADV}    
    tag = token[1][0]
    return tag_dict.get(tag, wordnet.NOUN)  # default when not found = 'n'

lemma_working_word_tokens = [lemmatizer.lemmatize(word) for word in working_word_tokens] 
tagged_lemma_working_word_tokens = [lemmatizer.lemmatize(token[0], pos=get_wordnet_pos(token)) for token in tagged_word_tokens]
 
print("Original words:", working_word_tokens)
print("Lemmas:", lemma_working_word_tokens)
print("Lemma with POS:", tagged_lemma_working_word_tokens)
```

### Combine Necessary Proprocessing Steps

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    import re, string
    from nltk.tokenize import word_tokenize 
    from nltk.corpus import stopwords 
    from nltk.stem import WordNetLemmatizer    
    
    word_tokens = word_tokenize(text)
    word_tokens = [word.lower() for word in word_tokens] 
    word_tokens = [re.sub(r'[^\w\s]', '', token) for token in word_tokens if re.sub(r'[^\w\s]', '', token)]
    corpus_stop_words = set(stopwords.words('english')) 
    word_tokens = [word for word in word_tokens if not word in corpus_stop_words] 
    lemmatizer = WordNetLemmatizer()
    word_tokens = [lemmatizer.lemmatize(word) for word in word_tokens]
    return word_tokens

text = 'Hello everyone. How are you? Today is a sunny day and better than yesterday.'
word_tokens = preprocess_text(text)
print(word_tokens)
```
