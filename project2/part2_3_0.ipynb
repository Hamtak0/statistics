```python
DATA_NAME_PATH = 'apparel/'
class_list = ['dress', 'pants', 'shirt', 'shoes']

IMG_SIZE = 224

DATASET_PATH = DATA_NAME_PATH + 'dataset/'
IMAGE_PATH = DATASET_PATH

print(DATASET_PATH)
```

### Load Dataset

```python
# Images saved in folders with the following patterns
# data_path/
#     train/
#        class0/
#           classs0_0.jpg
#           ...
#           classs0_k.jpg
#       class1/
#         classs1_0.jpg
#         ...
#         classs1_k.jpg
#       ...
#       classN/
#         classsN_0.jpg
#         ...
#         classsN_k.jpg
#     validation/
#        class0/
#        ...
#        classN/
#     test/
#        class0/
#        ...
#        classN/
# npy/
#     X_train.npy
#     y_train.npy
#     X_val.npy
#     y_val.npy
#     X_test.npy
#     y_test.npy
```

```python
train_ratio = 0.8
val_ratio = 0.10
test_ratio = 0.10
```

```python
class_names = class_list
class_names
```

```python
for split in ["train","validation","test"]:
    for cls in class_names:
        os.makedirs(os.path.join(DATASET_PATH, split, cls), exist_ok=True)
```

```python
import random
import shutil

for cls in class_names:

    src_folder = os.path.join(DATASET_PATH, cls)
    images = os.listdir(src_folder)

    random.shuffle(images)

    total = len(images)
    train_end = int(total * train_ratio)
    val_end = int(total * (train_ratio + val_ratio))

    train_imgs = images[:train_end]
    val_imgs = images[train_end:val_end]
    test_imgs = images[val_end:]

    for img in train_imgs:
        shutil.move(os.path.join(src_folder,img),
                    os.path.join(DATASET_PATH,"train",cls,img))

    for img in val_imgs:
        shutil.move(os.path.join(src_folder,img),
                    os.path.join(DATASET_PATH,"validation",cls,img))

    for img in test_imgs:
        shutil.move(os.path.join(src_folder,img),
                    os.path.join(DATASET_PATH,"test",cls,img))
```

```python
for cls in class_names:
    os.rmdir(os.path.join(DATASET_PATH, cls))
```

```python
import os
import cv2
import numpy as np
```

```python
NUMPY_PATH = DATA_NAME_PATH + 'npy/'
os.makedirs(NUMPY_PATH, exist_ok=True)
```

```python
# Read images and convert to numpy array
# with size = num_Images x IMG_SIZE x IMG_SIZE x 3 (RGB)

# Train folder
data = []
labels = []

TRAIN_PATH = IMAGE_PATH + 'train/'
# classes = os.listdir(TRAIN_PATH) # Fix random listdir bug

for label, category in enumerate(class_names):
    folder_path = os.path.join(TRAIN_PATH, category)
    for img in os.listdir(folder_path):

        img_path = os.path.join(folder_path, img)
        try:
            image = cv2.imread(img_path)                    # OpenCV loads images in BGR format
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR format to RGB format that deep-learning model requires
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

            data.append(image)
            labels.append(label)
        except:
            pass

# Convert to numpy
X_train = np.array(data)
y_train = np.array(labels)

# Normalize pixels
X_train = X_train / 255.0

print("Images shape:", X_train.shape)
print("Labels shape:", y_train.shape)
```

```python
np.save(NUMPY_PATH + "X_train.npy", X_train)
np.save(NUMPY_PATH + "y_train.npy", y_train)
```

```python
# Read images and convert to numpy array
# with size = num_Images x IMG_SIZE x IMG_SIZE x 3 (RGB)

# Validation folder
data = []
labels = []

VAL_PATH = IMAGE_PATH + 'validation/'
# classes = os.listdir(VAL_PATH)

for label, category in enumerate(class_names):
    folder_path = os.path.join(VAL_PATH, category)
    for img in os.listdir(folder_path):

        img_path = os.path.join(folder_path, img)
        try:
            image = cv2.imread(img_path)                    # OpenCV loads images in BGR format
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR format to RGB format that deep-learning model requires
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

            data.append(image)
            labels.append(label)
        except:
            pass

# Convert to numpy
X_val = np.array(data)
y_val = np.array(labels)

# Normalize pixels
X_val = X_val / 255.0

print("Images shape:", X_val.shape)
print("Labels shape:", y_val.shape)

np.save(NUMPY_PATH + "X_val.npy", X_val)
np.save(NUMPY_PATH + "y_val.npy", y_val)
```

```python
# Read images and convert to numpy array
# with size = num_Images x IMG_SIZE x IMG_SIZE x 3 (RGB)

# Test folder
data = []
labels = []

TEST_PATH = IMAGE_PATH + 'test/'
# classes = os.listdir(TEST_PATH)

for label, category in enumerate(class_names):
    folder_path = os.path.join(TEST_PATH, category)
    for img in os.listdir(folder_path):

        img_path = os.path.join(folder_path, img)
        try:
            image = cv2.imread(img_path)                    # OpenCV loads images in BGR format
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR format to RGB format that deep-learning model requires
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

            data.append(image)
            labels.append(label)
        except:
            pass

# Convert to numpy
X_test = np.array(data)
y_test = np.array(labels)

# Normalize pixels
X_test = X_test/ 255.0

print("Images shape:", X_test.shape)
print("Labels shape:", y_test.shape)

np.save(NUMPY_PATH + "X_test.npy", X_test)
np.save(NUMPY_PATH + "y_test.npy", y_test)
```
